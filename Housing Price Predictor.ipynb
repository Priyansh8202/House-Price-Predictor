```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42) 

#random_state=42 is use to fix the spliting values.if code run multiple time or any time, the split value will be same remain
```

```python
print(f"taining data set: {len(train_set)}\n Rows of test data set: {len(test_set)}")
```

```python
features = train_set.drop('Y house price of unit area',axis=1)
target = train_set['Y house price of unit area'].copy()
```

```python
#simpleImputer is use for kind of thing to automate like in below code. which is use to fill median values
# to all the cell if value of cells are empty

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy= 'median')
imputer.fit(train_set)
```

```python
imputer.statistics_
```

```python
X=imputer.transform(train_set)
train_set_tr = pd.DataFrame(X, columns=housing.columns)
```

## Creating a Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
mypipeline = Pipeline([('imputer', SimpleImputer(strategy='median')),
                       # ...add many as you want in your pipleline
                      ('std_sc', StandardScaler())])
housing_train_set = mypipeline.fit_transform(features)
```

```python
#By using pipeline we can perform as many as we want or and this is also known as automate the set of tasks.

#Standardization is use for convert all data coulumns into same sandard scale to follow 

housing_train_set
housing_train_set.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
#model = LinearRegression()
#model = DecisionTreeRegressor()
model = RandomForestRegressor()
model.fit(housing_train_set,target)
```

```python
some_data = features.iloc[:5]
some_target = target.iloc[:5]
predictor = mypipeline.transform(some_data)
model.predict(predictor)
```

```python
list(some_target)
```

```python
from sklearn.metrics import mean_squared_error
housing_predictions = model.predict(housing_train_set)
mse = mean_squared_error(target,housing_predictions)
rmse=np.sqrt(mse)
rmse
```

## Cross Validation of the Model
( For Batter Evalution Technique - Cross Validation )

Cross Validation is gernally use in the case of Underfitting OR Overfitting to Check the Mean Error of the 
model. If the model show's Mean Error (0.0) OR Errorless Prediction.

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(model, housing_train_set, target, scoring ='neg_mean_squared_error', cv=10)
rmse_score = np.sqrt(-score)
```

```python
rmse_score 
```

```python
def print_score(score):
    print("Score: ",score)
    print("Mean: ",score.mean())
    print("Score Deviation :",score.std())
```

```python
print_score(rmse_score)
```

## Testing Phase

```python
testX = test_set.drop('Y house price of unit area', axis=1)
testY = test_set['Y house price of unit area'].copy()
prepared_data = mypipeline.transform(testX)
final_predictions = model.predict(prepared_data)
final_mse = mean_squared_error(testY, final_predictions)
final_rmse = np.sqrt(final_mse)
final_rmse
```
